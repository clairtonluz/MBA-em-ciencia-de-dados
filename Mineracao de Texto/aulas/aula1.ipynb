# Stop words

```python
from spacy.lang.pt.stop_words import STOP_WORDS
```

```python
stop_words=STOP_WORDS
```

```python
len(stop_words)
```

```python
import string 
```

```python
pontuacoes=string.punctuation
pontuacoes
```

```python
pontuacoes=pontuacoes+"..."
```

# Criar uma funçao de processamento

```python
def processamento(texto):
    # texto em minusculo
    texto=texto.lower()
    documento=pln(texto)
    # remover stop words
    lista_tokens=[]
    for p in documento:
        if (p.text in stop_words)==False:
            lista_tokens.append(p)
    # remover pontuaçoes
    lista_tokens2=[]
    for p in lista_tokens:
        if (p.text in pontuacoes)==False:
            lista_tokens2.append(p)
    # lematizar
    lista_tokens3=[]
    for p in lista_tokens2:
        lista_tokens3.append(p.lemma_)
    
    return lista_tokens3
```
